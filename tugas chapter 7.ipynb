```python
import numpy as np # Mengimpor library komputasi numerik dasar NumPy

def integrate_rec(f, a, b, n): # Mendefinisikan fungsi integrasi numerik menggunakan metode Persegi Panjang (Rectangle Method)
    h = (b-a)/n # Menghitung lebar pias (step size) berdasarkan batas integrasi dan jumlah iterasi
    x = np.linspace(a, b, n+1)  # Menghasilkan array titik-titik diskrit pada sumbu X dengan jarak yang sama (equidistant)
    i = 0 # Menginisialisasi variabel counter untuk perulangan
    area = 0 # Menginisialisasi nilai awal luas area integrasi dengan nol
    while i < n: # Memulai perulangan untuk mengevaluasi luas setiap pias persegi panjang
        sup_rect =  f(x[i])*h # Menghitung luas satu persegi panjang (tinggi fungsi dikali lebar pias)
        area += sup_rect # Menambahkan luas pias saat ini ke total akumulasi area
        i += 1  # Melakukan iterasi ke pias berikutnya (increment)
    return area # Mengembalikan nilai total area hasil pendekatan numerik

sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5) # Memanggil fungsi integrasi untuk fungsi sinus pada interval [0, pi/2] dengan 5 iterasi (n=5)
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10) # Memanggil fungsi integrasi dengan 10 iterasi (n=10)
sup_100 = integrate_rec(np.sin, 0, np.pi/2, 100) # Memanggil fungsi integrasi dengan 100 iterasi (n=100) untuk akurasi lebih tinggi

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5)) # Mencetak output hasil estimasi n=5 dengan format presisi 2 desimal
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10)) # Mencetak output hasil estimasi n=10 dengan format presisi 2 desimal
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100)) # Mencetak output hasil estimasi n=100 dengan format presisi 2 desimal
```

```python
import numpy as np # Mengimpor modul NumPy untuk pemrosesan larik numerik

def integrate_trap(f, a, b, n): # Mendefinisikan fungsi kalkulasi integral tertentu menggunakan Aturan Trapesium (Trapezoidal Rule)
    h = (b-a)/n # Mengkalkulasi jarak antar titik (lebar interval/pias)
    x = np.linspace(a, b, n+1)  # Mendefinisikan titik-titik evaluasi sepanjang batas integrasi a hingga b
    i = 1 # Menetapkan indeks awal untuk iterasi (dimulai dari 1 karena indeks 0 sudah dievaluasi di inisialisasi area)
    area = h*(f(x[0]) + f(x[n]))/2 # Menginisialisasi komponen perhitungan batas awal dan batas akhir sesuai formulasi trapesium
    while i < n: # Melakukan iterasi untuk menghitung komponen bagian tengah kurva
        sup_rect = f(x[i])*h # Mengkalkulasi kontribusi dari titik interior ke dalam total luasan
        area += sup_rect  # Mengakumulasikan nilai estimasi ke dalam total area
        i += 1  # Meningkatkan nilai counter iterasi
    return area # Mengeluarkan nilai aproksimasi numerik akhir

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5) # Menjalankan uji integrasi aturan trapesium pada fungsi sinus dengan parameter n=5
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10) # Menjalankan uji integrasi aturan trapesium pada fungsi sinus dengan parameter n=10

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5)) # Mencetak hasil kalkulasi integrasi trapesium (n=5) ke layar
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10)) # Mencetak hasil kalkulasi integrasi trapesium (n=10) ke layar
```

```python
import numpy as np # Mengimpor NumPy untuk fungsionalitas array matematis
from scipy import integrate # Mengimpor sub-modul integrate dari library SciPy untuk algoritma kuadratur numerik

x = np.linspace(0, 9, 3) # Membentuk array diskrit batas domain dengan 3 titik (2 divisi/pias) dari nilai 0 hingga 9
y = x**2 # Membangun variabel dependen berdasarkan fungsi kuadrat eksponensial dari x

sup_trapz = np.trapezoid(y, x) # Melakukan integrasi numerik menggunakan fungsi metode trapezoidal bawaan NumPy versi pembaruan
sup_simps = integrate.simpson(y, x=x) # Melakukan integrasi numerik menggunakan metode Simpson bawaan SciPy versi pembaruan

print('Using n=2, the trapezoidal rule returns a value of {:.0f}'.format(sup_trapz)) # Mencetak hasil luasan kurva metode trapezoidal tanpa format desimal
print('Using n=2, the composite Simpson rule returns a value of {:.0f}'.format(sup_simps)) # Mencetak hasil luasan kurva metode Simpson tanpa format desimal
```

```python
import numpy as np # Memanggil library NumPy untuk operasi numerik array
from scipy import integrate # Memanggil utilitas integrasi numerik tingkat lanjut dari SciPy

conturs_areas = np.array([194135, 136366, 79745, 38335, 18450, 9635, 3895]) # Menyimpan data luasan kontur topografi ke dalam representasi NumPy array
x = np.array([0, 25, 50, 75, 100, 125, 150]) # Mendefinisikan array interval kedalaman/elevasi vertikal dengan spasiasi 25 meter

vol_traps = np.trapezoid(conturs_areas, x) # Mengestimasi volume geometri (integral luas terhadap elevasi) melalui pendekatan metode trapesium dari NumPy versi pembaruan
vol_simps = integrate.simpson(conturs_areas, x=x) # Mengestimasi volume geometri melalui algoritma integrasi metode Simpson dari pembaruan SciPy

print('The trapezoidal rule returns a volume of {:.0f} cubic meters'.format(vol_traps)) # Mencetak representasi volume estimasi metode trapesium dalam unit meter kubik
print('The composite Simpson rule returns a volume of {:.0f} cubic meters'.format(vol_simps)) # Mencetak representasi volume estimasi metode Simpson dalam unit meter kubik
```

```python
def simple_lithopress(z, ro=2900, g=9.8): # Mendeklarasikan fungsi kalkulasi tekanan litostatik dengan nilai default densitas batuan (ro) dan gravitasi (g)
    pressure = z * g * ro / 1e6 # Menghitung tekanan (P = rho*g*h) dan membaginya dengan 1.000.000 untuk konversi ke satuan MegaPascal (MPa)
    return pressure # Mengembalikan nilai skalar tekanan hasil perhitungan

my_pressure = simple_lithopress(z=2000) # Memanggil fungsi untuk mengalkulasi tekanan litostatik pada kedalaman (z) 2000 meter
print('The pressure at 2000 meters is {0:.0f} MPa'.format(my_pressure)) # Mencetak output tekanan dalam format bulat ke konsol pengguna
```

```python
import numpy as np # Mengimpor modul dasar NumPy
import matplotlib.pyplot as plt # Mengimpor perangkat visualisasi plot grafik

r = np.linspace(1, 6400, 6400) # Membuat representasi diskrit jarak radial interior bumi (1 km hingga 6400 km) sebanyak 6400 titik

def density(): # Membuat fungsi pembentukan profil stratifikasi densitas interior bumi (Earth's density profile)
    ro_inner_core = np.linspace(13100, 12800, 1220) # Mendefinisikan gradien densitas untuk zona Inti Dalam (Inner Core)
    ro_outer_core = np.linspace(12200, 9900, 2259) # Mendefinisikan gradien densitas untuk zona Inti Luar (Outer Core)
    ro_lower_mantle = np.linspace(5600, 4400, 2171) # Mendefinisikan gradien densitas untuk zona Mantel Bawah (Lower Mantle)
    ro_upper_mantle = np.linspace(4130, 3400, 720) # Mendefinisikan gradien densitas untuk zona Mantel Atas (Upper Mantle)
    ro_crust = np.linspace(3100, 2700, 30) # Mendefinisikan gradien densitas untuk zona Kerak Bumi (Crust)
   
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust)) # Menggabungkan (concatenate) semua lapisan menjadi satu array profil densitas utuh
    return ro_final # Mengembalikan array komprehensif profil densitas

ro = density() # Menjalankan fungsi dan memetakan output array ke variabel global 'ro'

fig, ax = plt.subplots() # Membangun area kerja kanvas dan set sumbu plot
ax.plot(r, ro, label="Densities on the Earth's Interior") # Menampilkan pemetaan visual antara jarak radial (r) terhadap profil densitas (ro)
ax.set_ylabel(r"Density [Kg/m$^3$]") # Menambahkan label ordinat (Y) dengan notasi satuan massa jenis
ax.set_xlabel("Distance from the Earth center r [km]") # Menambahkan label absis (X) sebagai parameter jarak dari pusat
ax.legend() # Memunculkan blok legenda untuk keterangan grafik
ax.grid() # Memunculkan jaring-jaring latar (grid) untuk mempermudah pembacaan kurva
plt.show() # Menginstruksikan modul untuk merender keluaran visual secara langsung
```

```python
def gravity(r): # Membangun definisi fungsi kalkulasi medan gravitasi interior (g) berdasarkan jarak jari-jari radial (r)
    
    g = np.zeros(len(r)) # Menyiapkan array kosong dengan panjang yang ekuivalen dengan r untuk menyimpan kalkulasi percepatan gravitasi
    Gr = 6.67408e-11 # Menginisialisasi Konstanta Gravitasi Universal (G)
    r = r * 1000 # Mengkonversi satuan array jarak parameter dari Kilometer (Km) menjadi satuan dasar Meter (m)
    
    for i in range(1, len(r)): # Membuka siklus perulangan untuk mengintegrasikan massa bumi dari pusat hingga titik kedalaman i
        
        r1 = r[0:i] # Melakukan pemotongan (slicing) batas array jari-jari evaluasi hingga posisi iterator
        ro1  = ro[0:i] # Melakukan pemotongan array densitas dari pusat hingga posisi iterator
        r2 = r1[i-1] # Mengidentifikasi limit jari-jari luar (shell radius) pada saat iterasi berjalan
    
        y = ro1 * r1**2 # Mengkalkulasi fungsi integran (rho * r^2) untuk mendapatkan nilai massa interior ekuivalen
        
        # PERBAIKAN: Menggunakan np.trapezoid bawaan NumPy
        y_int = np.trapezoid(y, r1) # Mengintegrasikan luasan kurva menggunakan kaidah numerik Trapezoidal (versi pembaruan) untuk estimasi parameter massa kumulatif
        
        g1 = ((4 * np.pi * Gr) / (r2**2)) * y_int # Menghitung percepatan gravitasi lokal (g) dengan menerapkan Hukum Gravitasi Newton
        g[i] = g1 # Menyimpan hasil gravitasi lokal (g1) pada titik tersebut ke dalam array penampung utama (g)
           
    return g # Mengembalikan hasil larik komplit perhitungan profil gravitasi bumi

g = gravity(r) # Mengeksekusi fungsi kalkulasi medan gravitasi menggunakan vektor radial r dan memuatnya ke variabel global g

fig, ax = plt.subplots() # Membuka instrumen figure untuk plotting data kalkulasi 
ax.plot(r, g) # Memetakan representasi titik percepatan gravitasi terhadap radius jarak dari inti
ax.grid() # Mengaktifkan matriks garis bantu skala sumbu (grid lines)
ax.set_ylabel(r'g [m/s$^2]$') # Mendefinisikan legenda teks sumbu Y menggunakan anotasi unit SI percepatan
ax.set_xlabel('Distance from the Earth center r [km]') # Mendefinisikan legenda teks sumbu X untuk koordinat spasial
plt.show() # Mengeksekusi rendering visual akhir plot distribusi gravitasi
```

```python
def pressure(r, ro, g): # Menyiapkan antarmuka fungsi untuk mengkalkulasi gradien tekanan hidrostatik/litostatik interior (P)
    
    p = np.zeros(len(r)) # Menginisialisasi matriks satu dimensi (vektor nol) untuk wadah hasil perhitungan tiap level kedalaman
    r = r * 1000 # Menyeragamkan unit kedalaman koordinat radial murni dari skema Kilometer menjadi meter (SI base unit)
    
    for i in range(0, len(r)): # Mendeklarasikan lup sekuensial untuk melakukan integrasi tekanan bertahap dari permukaan ke inti (arah turun)
        r1 = r[i:len(r)] # Mengambil set limit radius parsial (slicing vertikal) pada kolom overburden batuan di atas titik i
        ro1 = ro[i:len(r)] # Mengambil nilai densitas spesifik pada titik batas kalkulasi litostatik
        g1 = g[i:len(r)] # Mengambil profil percepatan gravitasi khusus yang bersesuaian dengan titik batas overburden
        y = ro1 * g1 # Menyusun fungsi perkalian komponen gaya berat per satuan volume (rho * g) sebagai integran tekanan
        
        # PERBAIKAN: Menggunakan np.trapezoid bawaan NumPy
        p1 = np.trapezoid(y, r1) # Menyelesaikan integral fungsi gaya berat terhadap lintasan jarak vertikal (r) via rule Trapesium numerik (versi pembaruan)
        
        p[i] = p1 # Mencatat hasil perhitungan kuadratur ke dalam sel matriks tekanan kumulatif (P)
    return p # Me-return larik total dari hasil pemetaan tekanan sepanjang model kedalaman bumi
    
p = pressure(r, ro, g) / 1e9 # Memanggil subrutin pressure dan langsung melakukan modifikasi skalar (pembagian 10^9) untuk konversi skala Pascal (Pa) menjadi GigaPascal (GPa)
z = np.linspace(6400, 1, 6400) # Membangkitkan sumbu koordinat referensi inversi kedalaman absolut (z=0 km ada di permukaan)

fig, ax = plt.subplots() # Mempersiapkan figur kerja Matplotlib untuk grafik tekanan litostatik interior
ax.plot(z, p) # Melakukan render plot grafis pemetaan kedalaman absolut (z) versus tekanan interior (P)
ax.grid() # Menampilkan grid lines referensi pada background figur area
ax.set_ylabel('P [GPa]') # Memasangkan metadata keterangan sumbu ordinat dengan unit GigaPascal
ax.set_xlabel('Depth z from the Earth Surface [km]') # Memasangkan metadata keterangan letak spasial (z) pada sumbu absis
plt.show() # Memicu mesin grafis internal untuk menghasilkan visualisasi akhir
```
